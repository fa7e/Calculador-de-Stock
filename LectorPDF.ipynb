```python
import pandas as pd

# Subir archivo
from google.colab import files
uploaded = files.upload()

# Leer el archivo de texto
filename = list(uploaded.keys())[0]
with open(filename, encoding="utf-8") as f:
    lines = [line.strip() for line in f if line.strip()]

# Filtrar solo líneas útiles
cleaned_lines = []
for line in lines:
    if not line.startswith("USD $") and "USD $" not in line:
        cleaned_lines.append(line)
    elif line.startswith("USD $"):
        cleaned_lines.append(line)

# Parsear líneas en bloques
items = []
i = 0
while i < len(cleaned_lines):
    try:
        item = cleaned_lines[i]
        material = cleaned_lines[i+1]
        product_line = cleaned_lines[i+2]
        price_str = cleaned_lines[i+3]
        subtotal_str = cleaned_lines[i+4]

        # Limpiar y convertir
        price = float(price_str.replace("USD $", "").replace(",", "").strip())
        subtotal = float(subtotal_str.replace("USD $", "").replace(",", "").strip())
        quantity = round(subtotal / price) if price else 0

        items.append([item, material, product_line, price, quantity, subtotal])
        i += 5
    except Exception as e:
        print(f"Error en línea {i}: {e}")
        i += 1

# Crear DataFrame
df = pd.DataFrame(items, columns=["Item", "Material Number", "Product Line", "Price (USD)", "Quantity", "Subtotal (USD)"])

# Total
total = df["Subtotal (USD)"].sum()
print(f"\n🧾 Total del carrito: USD ${total:,.2f}")

# Guardar Excel
df.to_excel("Shopping_Cart_Complete.xlsx", index=False)
files.download("Shopping_Cart_Complete.xlsx")
```
